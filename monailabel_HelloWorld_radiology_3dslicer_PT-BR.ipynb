## 2. Segmentação Automática com 3D Slicer

A seguir, apresentamos um tutorial passo a passo sobre o uso do 3D Slicer para segmentação automática do baço. Um modelo pré-treinado é fornecido e carregado pelo servidor MONAI Label.

### Instalação do 3D Slicer e do Plugin MONAI Label

Você pode baixar o 3D Slicer na [página de download do 3D Slicer](https://download.slicer.org/).
O MONAI Label é compatível com as versões estável e de pré-visualização do 3D Slicer: [Versão estável e de pré-visualização >= 5.0](https://download.slicer.org/).

**Para Ubuntu**

**Pré-requisitos para o 3D Slicer**: Espera-se que o Slicer funcione na grande maioria das distribuições Linux de desktop e servidor. O sistema deve fornecer pelo menos GLIBC 2.17 e GLIBCCC 3.4.19. Para mais detalhes, consulte os [pré-requisitos](https://slicer.readthedocs.io/en/latest/user_guide/getting_started.html#linux).

Como exemplo, os usuários do Ubuntu podem clicar no arquivo baixado do 3D Slicer, extraí-lo para a pasta de trabalho e depois clicar em "Slicer" para abrir o visualizador. Especificamente, abra o arquivo tar.gz e copie o diretório para o local de sua escolha. A instalação de pacotes adicionais pode ser necessária dependendo da distribuição e versão do Linux. Execute o executável do Slicer para iniciar o 3D Slicer.

**Para Windows**

Baixe e execute o instalador do 3D Slicer. Após a instalação, você pode executar o `Slicer` pelo Menu Iniciar do Windows.

#### Instalar o plugin MONAI Label pelo Gerenciador de Extensões do Slicer

**Passo 1:** Inicie o 3D Slicer e navegue até **View** -> **Extension Manager** -> **Active Learning** -> **MONAI Label**

<img src="https://raw.githubusercontent.com/Project-MONAI/tutorials/main/monailabel/figures/monailabel_radiology_3dslicer/3dslicer-extensions-manager.png" alt="drawing" width="600"/>

**Passo 2:** Instale o plugin MONAI Label (clique em "install") e reinicie o 3D Slicer. O botão "Restart" está no canto inferior direito da caixa de diálogo do Gerenciador de Extensões.

**Passo 3:** Adicionar atalho: Navegue até **Edit** -> **Application Settings**, no painel **Modules** arraste **MONAI Label** para o painel **Favorite Modules**. Clique em **OK**, reinicie o 3D Slicer, e o ícone do MONAI Label <img src= "https://raw.githubusercontent.com/Project-MONAI/tutorials/main/monailabel/figures/monailabel_radiology_3dslicer/MONAILabel.png" alt=" drawing" width=" 20"/> ficará visível na barra de ferramentas.

<img src="https://raw.githubusercontent.com/Project-MONAI/tutorials/main/monailabel/figures/monailabel_radiology_3dslicer/monai-label-plugin-favorite-modules-1.png" alt="drawing" width="600"/>


### 2.1 Iniciar o 3D Slicer e o Plugin MONAI Label

- Na barra de menus, clique em MONAI Label, conecte-se e verifique o endereço/nome do servidor MONAI Label, o aplicativo de radiologia, e confirme que o 3D Slicer está conectado corretamente ao servidor MONAI Label. Se não estiver, preencha o endereço e a porta corretos do servidor.

![slicer1](https://raw.githubusercontent.com/Project-MONAI/tutorials/main/monailabel/figures/monailabel_radiology_3dslicer/slicer1.png)


### 2.2 Selecionar Modelo e Buscar Dados

- Selecione o modelo que deseja usar, como "**segmentation_spleen**" neste tutorial, clique em **next sample** para carregar os dados de imagem do repositório de estudos.

![slicer1](https://raw.githubusercontent.com/Project-MONAI/tutorials/main/monailabel/figures/monailabel_radiology_3dslicer/slicer2.png)


### 2.3 Executar a Segmentação Automática

- O MONAI Label fornece pesos de modelo pré-treinados para o modelo de segmentação de baço. Clique em **run** para realizar a inferência nos dados atualmente carregados.

- Nota: os modelos pré-treinados serão baixados ao iniciar o servidor MONAI Label na pasta "**model**". Por exemplo, neste caso de uso, o modelo pré-treinado baixado é salvo em "**apps/radiology/model/pretrained_segmentation_spleen.pt**".

- Os usuários podem mover os cortes e a localização para visualizar a segmentação de inferência automática.

![slicer1](https://raw.githubusercontent.com/Project-MONAI/tutorials/main/monailabel/figures/monailabel_radiology_3dslicer/slicer3.png)


### 2.4 Editar Anotações e Enviar Rótulo

- Os usuários podem editar os rótulos de inferência manualmente usando o Editor de Segmentos, a ferramenta para ajustar anotações. Após concluir a anotação, clique em **Submit Label** para salvar o rótulo de referência no sistema de arquivos.

- Nota: as anotações finais serão salvas na pasta "**labels/final**" no conjunto de dados de estudo. Por exemplo, neste caso de uso, o rótulo de referência será salvo em "**datasets/Task09_Spleen/imagesTr/labels/final**".

## 3. Aprendizado Ativo com 3D Slicer

O aprendizado ativo e o ajuste fino interativo de modelos são recursos destacados do MONAI Label. Os usuários podem treinar seus modelos a qualquer momento quando novos rótulos anotados forem salvos. Clique no botão **Train**; o servidor MONAI Label buscará o rótulo de referência final salvo e fará o ajuste fino do modelo anterior.

![slicer1](https://raw.githubusercontent.com/Project-MONAI/tutorials/main/monailabel/figures/monailabel_radiology_3dslicer/slicer5.png)

Os usuários podem realizar a segmentação automática usando o modelo mais recente com ajuste fino.
O processo de aprendizado ativo selecionará lotes de dados não rotulados.

Ao selecionar a próxima imagem, as imagens já treinadas serão marcadas como dados rotulados e não serão selecionadas para o próximo lote de rotulagem. Estratégias de aprendizado ativo como "first/random" serão usadas para selecionar quais dados não rotulados buscar.

Para o design geral do fluxo de trabalho de aprendizado ativo, veja o slide: ![activeL](https://raw.githubusercontent.com/Project-MONAI/tutorials/main/monailabel/figures/monailabel_radiology_3dslicer/activeLearning.png)


## 4. Repetir o Processo de Rotulagem Interativa

Repita o processo de busca de dados e aprendizado ativo a partir da seção 2.2, até que todos os dados não rotulados estejam anotados e treinados.

Para cada ciclo de treinamento, o novo modelo com melhor métrica será salvo em "**model/<model_name>.pt**". Neste caso de uso, "**apps/radiology/model/segmentation_spleen.pt**" é salvo. Para estatísticas detalhadas de treinamento e modelos por etapa, consulte "**apps/radiology/model/segmentation_spleen/train_01**".
