## 2. Segmentación Automática con 3D Slicer

A continuación se presenta un tutorial paso a paso sobre el uso de 3D Slicer para la segmentación automática del bazo. El servidor MONAI Label proporciona y carga un modelo preentrenado.

### Instalación de 3D Slicer y el plugin MONAI Label

Puede descargar 3D Slicer en la [página de descarga de 3D Slicer](https://download.slicer.org/).
MONAI Label es compatible con las versiones estable y de vista previa de 3D Slicer: [Versión estable y de vista previa >= 5.0](https://download.slicer.org/).

**Para Ubuntu**

**Requisitos previos para 3D Slicer**: Se espera que Slicer funcione en la gran mayoría de distribuciones Linux de escritorio y servidor. El sistema debe proporcionar al menos GLIBC 2.17 y GLIBCCC 3.4.19. Para más detalles, consulte los [requisitos previos](https://slicer.readthedocs.io/en/latest/user_guide/getting_started.html#linux).

Como ejemplo, los usuarios de Ubuntu pueden hacer clic en el archivo descargado de 3D Slicer, extraerlo en la carpeta de trabajo y luego hacer clic en "Slicer" para abrir el visor. Específicamente, abra el archivo tar.gz y copie el directorio a la ubicación de su elección. Es posible que sea necesario instalar paquetes adicionales según la distribución y versión de Linux. Ejecute el ejecutable de Slicer para iniciar 3D Slicer.

**Para Windows**

Descargue y ejecute el instalador de 3D Slicer. Una vez instalado, puede ejecutar `Slicer` desde el menú Inicio de Windows.

#### Instalar el plugin MONAI Label desde el Administrador de Extensiones de Slicer

**Paso 1:** Inicie 3D Slicer y navegue a **View** -> **Extension Manager** -> **Active Learning** -> **MONAI Label**

<img src="https://raw.githubusercontent.com/Project-MONAI/tutorials/main/monailabel/figures/monailabel_radiology_3dslicer/3dslicer-extensions-manager.png" alt="drawing" width="600"/>

**Paso 2:** Instale el plugin MONAI Label (haga clic en "install") y reinicie 3D Slicer. El botón "Restart" se encuentra en la esquina inferior derecha del cuadro de diálogo del Administrador de Extensiones.

**Paso 3:** Agregar acceso directo: Navegue a **Edit** -> **Application Settings**, en el panel **Modules** arrastre **MONAI Label** al panel **Favorite Modules**. Haga clic en **OK**, reinicie 3D Slicer, y el ícono de MONAI Label <img src= "https://raw.githubusercontent.com/Project-MONAI/tutorials/main/monailabel/figures/monailabel_radiology_3dslicer/MONAILabel.png" alt=" drawing" width=" 20"/> será visible en la barra de herramientas.

<img src="https://raw.githubusercontent.com/Project-MONAI/tutorials/main/monailabel/figures/monailabel_radiology_3dslicer/monai-label-plugin-favorite-modules-1.png" alt="drawing" width="600"/>


### 2.1 Iniciar 3D Slicer y el Plugin MONAI Label

- En la barra de menú, haga clic en MONAI Label, conéctese y verifique la dirección/nombre del servidor MONAI Label, la aplicación de radiología, y confirme que 3D Slicer esté conectado correctamente al servidor MONAI Label. Si no es así, ingrese la dirección y el puerto correctos del servidor.

![slicer1](https://raw.githubusercontent.com/Project-MONAI/tutorials/main/monailabel/figures/monailabel_radiology_3dslicer/slicer1.png)


### 2.2 Seleccionar Modelo y Obtener Datos

- Seleccione el modelo que desea usar, como "**segmentation_spleen**" en este tutorial, haga clic en **next sample** para cargar los datos de imagen desde el almacén de estudios.

![slicer1](https://raw.githubusercontent.com/Project-MONAI/tutorials/main/monailabel/figures/monailabel_radiology_3dslicer/slicer2.png)


### 2.3 Ejecutar la Segmentación Automática

- MONAI Label proporciona pesos de modelo preentrenados para el modelo de segmentación de bazo. Haga clic en **run** para realizar la inferencia sobre los datos cargados actualmente.

- Nota: los modelos preentrenados se descargarán al iniciar el servidor MONAI Label en la carpeta "**model**". Por ejemplo, en este caso de uso, el modelo preentrenado descargado se guarda en "**apps/radiology/model/pretrained_segmentation_spleen.pt**".

- Los usuarios pueden mover los cortes y la ubicación para ver la segmentación de inferencia automática.

![slicer1](https://raw.githubusercontent.com/Project-MONAI/tutorials/main/monailabel/figures/monailabel_radiology_3dslicer/slicer3.png)


### 2.4 Editar Anotaciones y Enviar Etiqueta

- Los usuarios pueden editar las etiquetas de inferencia manualmente usando el Editor de Segmentos, la herramienta para ajustar anotaciones. Una vez finalizada la anotación, haga clic en **Submit Label** para guardar la etiqueta de verdad fundamental en el sistema de archivos.

- Nota: las anotaciones finales se guardarán en la carpeta "**labels/final**" del conjunto de datos de estudio. Por ejemplo, en este caso de uso, la etiqueta de verdad fundamental se guardará en "**datasets/Task09_Spleen/imagesTr/labels/final**".

## 3. Aprendizaje Activo con 3D Slicer

El aprendizaje activo y el ajuste fino interactivo de modelos son características destacadas de MONAI Label. Los usuarios pueden entrenar sus modelos en cualquier momento cuando se guarden nuevas etiquetas anotadas. Haga clic en el botón **Train**; el servidor MONAI Label obtendrá la etiqueta de verdad fundamental final guardada y ajustará finamente el modelo anterior.

![slicer1](https://raw.githubusercontent.com/Project-MONAI/tutorials/main/monailabel/figures/monailabel_radiology_3dslicer/slicer5.png)

Los usuarios pueden realizar la segmentación automática usando el modelo ajustado más reciente.
El proceso de aprendizaje activo seleccionará datos por lotes sin etiquetar.

Al seleccionar la siguiente imagen, las imágenes ya entrenadas se marcarán como datos etiquetados y no se seleccionarán para el siguiente lote de etiquetado. Se usarán estrategias de aprendizaje activo como "first/random" para seleccionar qué datos sin etiquetar obtener.

Para el diseño general del flujo de trabajo de aprendizaje activo, consulte la diapositiva: ![activeL](https://raw.githubusercontent.com/Project-MONAI/tutorials/main/monailabel/figures/monailabel_radiology_3dslicer/activeLearning.png)


## 4. Repetir el Proceso de Etiquetado Interactivo

Repita el proceso de obtención de datos y aprendizaje activo desde la sección 2.2, hasta que todos los datos sin etiquetar estén anotados y entrenados.

Para cada ciclo de entrenamiento, el nuevo modelo con mejor métrica se guardará en "**model/<model_name>.pt**". En este caso de uso, se guarda "**apps/radiology/model/segmentation_spleen.pt**". Para estadísticas detalladas de entrenamiento y modelos por paso, consulte "**apps/radiology/model/segmentation_spleen/train_01**".
